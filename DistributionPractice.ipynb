```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

```python
grades = pd.read_csv('data/StudentsPerformance.csv')
grades.sort_values(by=['math score'], inplace=True)

x = grades['math score'].values
```

```python
grades.head()
```

```python
# Number of data points
N = len(x)

# ---------------- CDF ----------------
# Step 1: Compute cumulative probabilities
y_cdf = x/len(x)

# Step 2: Plot CDF
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)  # 1 row, 2 cols, plot 1
plt.step(x, y_cdf, where='post')

plt.title("CDF of Math Grades")
plt.xlabel("Grade")
plt.ylabel("Cumulative Probability")
plt.grid(True)

# ---------------- PDF ----------------
# Step 1: Compute histogram for PDF
pdf_y, pdf_x = np.histogram(x, bins=10, density=True)

# Step 2: Plot PDF as bar plot
plt.subplot(1,2,2)  # 1 row, 2 cols, plot 2
plt.bar(pdf_x[:-1], pdf_y, width=np.diff(pdf_x), edgecolor='black', align='edge')
plt.title("PDF of Math Grades")
plt.xlabel("Grade")
plt.ylabel("Probability Density")
plt.grid(True)

#plt.tight_layout()
plt.show()
```
